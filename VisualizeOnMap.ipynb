```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mplleaflet
```

```python
data = pd.read_csv( './train.csv' )
geoData = pd.read_csv( './train_lat_lon.csv' )
```

```python
geoData.head()
```

```python
# Объединяем по идентификатору
joinedData = data.join( geoData.set_index('id'), on = 'id')
```

```python
# Берем 1% от выборки, иначе заканчивается память.
skippedData = joinedData[::100]
```

```python
# Размер маркера пропорционален цене сделки.
plt.rcParams["figure.figsize"] = (10, 10)
ax = skippedData.plot.scatter( x = 'lon', y = 'lat', s = skippedData.price_doc / 1e5 )
mplleaflet.display( fig = ax.figure, closefig = True )
```
