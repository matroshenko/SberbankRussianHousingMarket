```python
import numpy as np
import pandas as pd
import datetime as dat
import calendar
import math
```

```python
Index_RE = pd.read_csv("./index_re.csv", sep=';', parse_dates=['Time'])
cpi = pd.read_csv("./CPI.csv", sep=';', parse_dates=['Time'])
salary = pd.read_csv("./salary.csv", sep=';')
macro_data = pd.read_csv("./macro.csv", parse_dates=['timestamp'], 
                                    usecols=['timestamp'])
```

```python
#Setting up observation lags
Real_estate_lag_q = 1
CPI_lag = 2
Salary_lag = 1
```

```python
#Setting columns for merge
macro_data['year'] = macro_data['timestamp'].dt.year
macro_data['month']=macro_data['timestamp'].dt.year*12 + macro_data['timestamp'].dt.month
macro_data['month_q']=macro_data['month'] - macro_data['timestamp'].dt.month % 3

Index_RE['month_q']=Index_RE['Time'].dt.year*12 + Index_RE['Time'].dt.month + Real_estate_lag_q*3
cpi['month'] = cpi['Time'].dt.year*12 + cpi['Time'].dt.month + CPI_lag
salary['year'] = salary['Time'] + Salary_lag
```

```python
cpi=cpi.drop('Time',axis=1)
salary=salary.drop('Time',axis=1)
Index_RE=Index_RE.drop('Time',axis=1)
```

```python
macro_data = pd.merge_ordered(cpi, macro_data, on='month', how='right')
macro_data = pd.merge_ordered(salary, macro_data, on='year', how='right')
macro_data = pd.merge_ordered(Index_RE, macro_data, on='month_q', how='right')
```

```python
macro_data = macro_data.dropna(how= 'any',axis = 0)
macro_data =  macro_data.drop('year', 1)
macro_data =  macro_data.drop('month', 1)
macro_data =  macro_data.drop('month_q', 1)
```

```python
macro_data.to_csv( './macro1.csv', index = False )
```
